```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style(rc={"axes.facecolor":"#F0F8FF"})
raw_data=pd.read_csv("Seaborndp.csv")
raw_data
print(raw_data.shape)
raw_data.head()
```

```python
#Example 1 - Simple Line Graph
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
ax=sns.lineplot(x="Week_ID",y="Revenue",data=raw_data)
sns.set()
```

```python
#Example 2 - Adding Categories
ax=sns.lineplot(x="Week_ID",y="Revenue",hue="Promo",data=raw_data)
```

```python
#Example 3 - By Promo with Style
ax=sns.lineplot(x="Week_ID",y="Revenue",hue="Promo",style="Promo",data=raw_data)
```

```python
#Example 4 - By promo with style & increase the size & remove error bars
#ci=Confidence Interval
#Example 1 - simple line graph
#Assumin we want to investigate the Revenue by Date
#ci= confidence interval
#ci computes the percentile limits...Actual mean, sd in the percentile routine
#increase the size
sns.set(rc={'figure.figsize':(12,10)})
ax=sns.lineplot (x='Week_ID',y='Revenue',hue='Promo',style='Promo',data=raw_data,ci=None)
```

```python
sns.set(rc={'figure.figsize':(12,10)})
ax=sns.lineplot(x='Week_ID',y='Revenue',hue='Promo',style='Promo',data=raw_data,ci='sd')
```

```python
#Example 5 - By Promo with style and Increase the size and Remove th error bars and adding markers & by month
ax=sns.lineplot(x='Month_ID',y='Revenue',hue='Promo',style='Promo',data=raw_data,ci=None,markers=True)
```

```python
#Example 6 - By Day_Name with style & Increase the Size and Remove error bars & adding markers by month
ax=sns.lineplot(x='Month_ID',y='Revenue',hue='Day_Name',style='Promo',data=raw_data,ci=None,markers=True)
```

# Bar Plots

```python
#Example 1 - Total Revenue by Month
#Notes:
#1 - the lines signify the ci
#2 - takes mean by default
ax=sns.barplot(x="Month_ID",y="Revenue",data=raw_data) #hue="Month_ID", ci=None
#as_index = False. agg({'Revenue':'mean'})
raw_data[['Month_ID','Revenue']].groupby('Month_ID',as_index = False).agg({'Revenue':'mean'})


```

```python
#Examle 2 - Total Revenue by Month - Remove ci
ax=sns.barplot(x="Month_ID",y="Revenue",data=raw_data,ci=False)
```

```python
#Example 3 - Total Revenue by Month - Remove the ci - By Promo
#group by clause ---> hue
#Three different category under one ID...
ax=sns.barplot (x="Month_ID",y="Revenue",data=raw_data,ci=True,hue='Promo')
```

```python
#Example 3 - Total Revenue by Month- Remove the confidence Interval - By Promo
ax=sns.barplot(x="Month_ID",y="Revenue",data=raw_data,ci=False,hue="Promo")
```

```python
#Example 3 - Total Revenue by Month- Add sd the confidence Interval - By Promo
ax=sns.barplot(x="Month_ID",y="Revenue",data=raw_data,ci='sd',hue="Promo")
```

```python
#Example 4 - Total Revenue by Month- Remove the confidence Interval - By Promo - changing direction
ax=sns.barplot(x="Revenue",y="Year",data=raw_data,ci=False,hue="Promo",orient='h')
```

```python
#Example 5 - Total Revenue by Month- Remove the confidence Interval - By Promo - changing direction - chaning color
ax=sns.barplot(x="Revenue",y="Year",data=raw_data,ci=False,hue="Promo",orient='h',color='#3359FF')
```

# Histograms

```python
#Example 1 investigating the distribution
x=raw_data['Revenue'].values
sns.distplot(x,color='red')
```

```python
#Example 2 Investigating the distribution of revenue, adding the mean
x=raw_data['Revenue'].values
sns.distplot(x,color='blue')
#calculating the mean
mean=raw_data['Revenue'].mean()
#plotting the mean
plt.axvline(mean,0,1,color='red')
```

```python
#Example 2 Investigating the distribution of visitors, adding the mean
x=raw_data['Visitors'].values
sns.distplot(x,color='#FF9C33');
#Calculating the mean
mean= raw_data['Visitors'].mean()
#plotting the mean
plt.axvline(mean,0,1,color='blue')
```

# MEAN, MEDIAN, MODE

```python
import numpy as np
incomes=np.random.normal(25000,15000,100000)
m=np.mean(incomes)
print(m)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.hist(incomes,50)
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.hist(incomes,100)
plt.show()
```
