```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set_style(rc={"axes.facecolor":"lightsteelblue"})
sns.set()
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
may_flights=flights.query("month=='May'")
sns.lineplot(data=may_flights,x="year",y="passengers")
sns.set()
```

```python
flights_wide=flights.pivot("year","month","passengers")
flights_wide.head()
```

```python
sns.lineplot(data=flights_wide["May"])
sns.set()
```

```python
sns.lineplot(data=flights_wide)
sns.set()
```

```python
sns.set_style(rc={"axes.facecolor":"#F0F8FF"})
sns.lineplot(data=flights,x="year",y="passengers")
sns.set()
```

```python
sns.set_style(rc={"axes.facecolor":"#FFEFDB"})
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
sns.set()
```

```python
sns.set_style(rc={"axes.facecolor":"#C1CDCD"})
sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month")
sns.set()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#sns.set_theme(style="darkgrid")
plt.style.use("dark_background")
sns.dark_palette("#79C")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
sns.set_style(rc={"axes.facecolor":"#FFD39B"})
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event", style="event",markers=True, dashes=False)
sns.set()
```

# Show error bars instead of error bands and extend them to two standard error widths:

```python
#Error bars instead of error bands 68% confidence interval 
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", err_style="bars", ci=68)
sns.set()
```

```python
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
sns.set()
```

# Assigning the units variable will plot multiple lines without applying a semantic mapping:

```python
sns.lineplot(data=fmri.query("region == 'frontal'"),x="timepoint", y="signal", hue="event", units="subject",
             estimator=None, lw=.4,)
sns.set()
```

```python
sns.lineplot(data=fmri.query("region == 'frontal'"),x="timepoint", y="signal", hue="event", units="subject",estimator=None,)
sns.set()
```
