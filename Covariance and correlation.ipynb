# Covariance and correlation
Covariance measures how two variabls vary in tandem from thier means.
For example, let's say we work for an e-commerce company and they are interested in finding a correlation between page speed and how much a customer spends.
Basically we treat each varaible as a vector of deviations from their mean and compute the dot product of both vectors. Geometrically this can be thought of as the angle between the two vectors in a high dimensional space but you can just think of it as a measure of similarity between the two variables.

```python
%matplotlib inline
import numpy as np
from pylab import *

#Step- 2
def de_mean(x):
    xmean=mean(x)
    return [xi - xmean for xi in x]

#Step- 3

def covariance (x,y):
    n=len(x)
    return dot(de_mean(x),de_mean(y))/(n-1) #Sample Variance

#page loading time
Pagespeeds = np.random.normal(3.0,1.0,1000)
#Customer Purchase
purchaseAmount = np.random.normal(50.0,10.0,1000) #($,dev,data)
scatter(Pagespeeds,purchaseAmount)
covariance(Pagespeeds,purchaseAmount)
```

# 
Now we'll make our fabricated purchase amount an actual function of page speed, making a very real correlation. The negative value indicates an inverse relationship; pages that render in less time result in more money spent.

```python
#fabricate or cleaning our data
#Inverse Proportion
purchaseAmount=np.random.normal(50.0,10.0,1000) / Pagespeeds
scatter(Pagespeeds, purchaseAmount)
covariance(Pagespeeds,purchaseAmount)
```

# 
Covariance is sensitive to the units in the variables, which makes it difficult to interpret. Correlation normalizes everything by their standard deviations, giving an easier to understand value that ranges from -1 (for a perfect inverse correlation) to 1 (perfect positive correlation)

```python
def correlation(x,y):
    stddevx=x.std()
    stddevy=y.std()
    return covariance (x,y) / stddevx / stddevy
correlation(Pagespeeds, purchaseAmount)
```

```python
np.corrcoef(Pagespeeds,purchaseAmount)
```
