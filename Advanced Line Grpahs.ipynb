# Bar Plots using matplotlib & Numpy

```python
#Example 1
category=['Production A','Production B','Production C']
revenue=np.array([500,400,800])
plt.figure(figsize=(5,5))
plt.bar(range(0,3),revenue)
plt.xticks(range(0,3),category)
plt.show()
```

```python
#Example 2 - subplots
category=['Account A','Account B','Account C','Account D']
revenue=np.array([200,400,100,700])
plt.figure(figsize=(15,5))
#bar
plt.subplot(131)
plt.bar(range(0,4),revenue)
plt.xticks(range(0,4),category)
plt.grid(True)

#line
plt.subplot(132)
plt.plot(range(0,4),revenue,c='red')
plt.xticks(range(0,4),category)
plt.grid(True)

#scatter
plt.subplot(133)
plt.scatter(range(0,4),revenue,c='green')
plt.xticks(range(0,4),category)

plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

raw_data=pd.read_csv("Matplotlibdata.csv")
```

```python
#example 6 - combo stack bars
a=raw_data.tail(10).reset_index()
plt.figure(figsize=(15,10))
plt.bar(range(0,len(a["Revenue"])),a["Revenue"],label="Revenue",linewidth=5.,color="green")
plt.bar(range(0,len(a["Visitors"])),a["Marketing Spend"],label="Marketing Spend",linewidth=5.,color="red")
plt.plot(a["Visitors"],linestyle="-.")
plt.xticks(range(0,len(a["Visitors"])),a["Date"],rotation=90)
plt.title("Visitors & Revenue by Day",fontsize=20)
plt.legend(loc="upper right",fontsize=14)
plt.xlabel("Dates")
plt.grid(True)
plt.show()
```

```python
#example 7 - Combp 2+ bar graphs + Secondary Scale
weight=.30
a[["Revenue","Marketing Spend"]].plot(kind="bar",color=["green","red"],figsize=(15,10))
a["Visitors"].plot(c="blue",linestyle="-.",secondary_y=True)
plt.xticks(range(0,len(a["Visitors"])),a["Date"],rotation=90)
plt.title("Visitors & Revenue by Day",fontsize=20)
plt.legend(["Visitors"],loc="upper left",fontsize=14)
plt.xlabel("Dates")
plt.xlim(-weight,len(a["Visitors"])-weight)
plt.show()
```

# Scatterplot using DFs

```python
#Example 1 - Simple scatterplot
plt.figure(figsize=(15,10))
plt.scatter(raw_data["Revenue"],raw_data["Marketing Spend"],c="red",marker="+")
plt.title("Relationship between M.Spend & Revenue",fontsize=20)
plt.legend(["Marketing Spend"],loc="upper left",fontsize=14)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")
plt.grid(True)
plt.show()
```

```python
groups=raw_data[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
groups
fig,ax=plt.subplots(figsize=(15,10))
for promo, group in groups:
    ax.scatter(group["Revenue"],group["Marketing Spend"],marker="x",label=promo)

ax.legend(["Promo","Revenue","Marketing Spend"],loc="upper left",fontsize=14)
plt.title("Relationship between M.Spend & Revenue by Promo",fontsize=20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")
plt.grid(True)
plt.show()
```

```python
#Example 3 - scatterplot with groups
groups=raw_data[["Day_Name","Revenue","Marketing Spend"]].groupby("Day_Name")
groups
fig,ax=plt.subplots(figsize=(15,10))
for promo,group in groups:
    ax.scatter(group['Revenue'],group["Marketing Spend"],marker="x",label=promo)

ax.legend(loc="upper left",fontsize=14)
plt.title("relationship between M.Spend & Revenue by Day_Name",fontsize=20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")
plt.grid(True)
plt.show()
```
