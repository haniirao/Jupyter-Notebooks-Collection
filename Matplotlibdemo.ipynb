# use Numpy's linspace function to create an array x containing 500 floats ranging from -2 to 2, then we create a second array y computed as the square of x.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y= x**2
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("X")
plt.ylabel("Y=x**2")
plt.grid(True)#color='r',linestyle='-',linewidth=2
plt.show()
```

```python
x=np.linspace(-1.4,1.4,30)
line1,line2,line3=plt.plot(x,x,'g--',x,x**2,'r:',x,x**3,'b^')
line1.set_linewidth(3.0)
line2.set_dash_capstyle('round') #shape of the line
line3.set_alpha(0.5) #depth of the line
plt.show()
```

# Saving a Figure
Saving a figure is as simple as calling savefig with the name of the file

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-1.4,1.4,30)
plt.plot(x,x**2)
plt.savefig(r"C:\Users\HP\Desktop\ANALYTICS\square_function.png",transparent=False)
plt.show()
```

```python
x=np.linspace(-1.4,1.4,30)
line1,line2,line3=plt.plot(x,x,'g--',x,x**2,'r:',x,x**3,'b^')
line1.set_linewidth(3.0)
line2.set_dash_capstyle('round') #shape of the line
line3.set_alpha(0.5) #depth of the line
plt.savefig(r"C:\Users\HP\Desktop\ANALYTICS\functions.png",transparent=True)
plt.show()
```

```python
#Correct Version
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

bar_plot =  ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
# print(type(bar))

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(handles=bar_plot, labels=bar_labels, title='Fruit color')

plt.savefig(r"C:\Users\HP\Desktop\ANALYTICS\fruit supply.png",transparent=False)

plt.show()
```
